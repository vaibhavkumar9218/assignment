```python
from tkinter import *
from tkinter import messagebox

# Define the initial balances and passwords for each account
accounts = {"123451": {"balance": 7000, "password": "1231"},
            "123452": {"balance": 6200, "password": "1232"},
            "123453": {"balance": 98100, "password": "1233"},
            "123454": {"balance": 50200, "password": "1234"},
            "234565": {"balance": 103600, "password": "5678"}}

# Define a function to update the balance
def update_balance(amount):
    global balance
    balance += amount
    balance_label.config(text=f"Current Balance: {balance}$")

# Define a function to withdraw money
def withdraw_money():
    global balance
    amount = int(amount_entry.get())
    if amount <= balance:
        update_balance(-amount)
        messagebox.showinfo("Success", f"Withdrawal of {amount}$ successful")
    else:
        messagebox.showerror("Error", "Insufficient balance")

# Define a function to deposit money
def deposit_money():
    global balance
    amount = int(amount_entry.get())
    update_balance(amount)
    messagebox.showinfo("Success", f"Deposit of {amount}$ successful")

# Define a function to check the password and select the account
def check_password():
    global balance, password
    account_name = account_entry.get()
    if account_name in accounts:
        password = accounts[account_name]["password"]
        if password_entry.get() == password:
            balance = accounts[account_name]["balance"]
            balance_label.config(text=f"Current Balance: {balance}$")
            amount_entry.config(state="normal")
            withdraw_button.config(state="normal")
            deposit_button.config(state="normal")
        else:
            messagebox.showerror("Error", "Password incorrect")
    else:
        messagebox.showerror("Error", "Account not found")

# Define a function to exit the program
def exit_program():
    root.destroy()

# Create the main window
root = Tk()
root.title("ATM Machine")
root.configure(bg="#68fc29")

# Create the widgets
welcome_label = Label(root, text="Welcome to the ATM Machine", font=("Arial", 50), bg="#f935da")
account_label = Label(root, text="Enter account number:", bg="#55e219")
account_entry = Entry(root, font=("Arial", 16))
password_label = Label(root, text="Enter password:", bg="#55e219")
password_entry = Entry(root, show="*", font=("Arial", 16))
password_button = Button(root, text="Enter", font=("Arial", 16), command=check_password)
balance_label = Label(root, text="", font=("Arial", 16), bg="#55e219")
amount_label = Label(root, text="Enter amount:", bg="#55e219")
amount_entry = Entry(root, font=("Arial", 16), state="disabled")
withdraw_button = Button(root, text="Withdraw", font=("Arial", 16), command=withdraw_money, state="disabled")
deposit_button = Button(root, text="Deposit", font=("Arial", 16), command=deposit_money, state="disabled")
exit_button = Button(root, text="Exit", font=("Arial", 16),bg="#55e219", command=exit_program)

# Add the widgets to the window
welcome_label.pack(pady=20)
account_label.pack(pady=10)
account_entry.pack(pady=10)
password_label.pack(pady=10)
password_entry.pack(pady=10)
password_button.pack(pady=10)
balance_label.pack(pady=20)
amount_label.pack(pady=10)
amount_entry.pack(pady=10)
withdraw_button.pack(pady=10)
deposit_button.pack(pady=10)
exit_button.pack(pady=10)

# Run the main loop
root.mainloop()
```
